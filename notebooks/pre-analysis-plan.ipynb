### Abstract

This project attempts to reproduce analysis by Chris Pyles on a data set of insurance charges from Kaggle. The original analysis conducted an A/B Test on the smoking and non-smoking populations in an attempt to determine whether or not being a smoker influences your insurance premia. In our recreation of this analysis, we will attempt to reproduce the A/B Test results and comment on statistical errors and unconsidered problems that the original analysis did not cover. We will reproduce the A/B Test in two ways: including and excluding outliers in insurance charges, something that the original analysis fails to do. We also comment on the reliability of the data set, given that it is from Kaggle and may not be from a reliable source.

### Data

The data is from the [Github repository](https://github.com/chrispyles/jupyter/blob/master/content/notebooks/insurance/insurance.csv) containing the original analysis, and it came from Kaggle before that. Research into the data shows that its original provenance is from _Machine Learning for R_ by Brett Lantz. It is unclear how this data was collected, but its original purpose was to illustrate a machine learning concept for an exercise in Lantz's text. It is a CSV file and there appear to be no missing values.

### Strategy

The conding analysis will be performed in two versions, one using the `datascience` library and the other using `pandas`. To conduct this portion of our analysis, we will load the data from the CSV file and first explore the data. This includes categorizing variables, plotting relevant numerical data, and considering core statistics of the data we are interested in. We then move on to the first of two A/B Tests, which we will conduct by shuffling the smoker and non-smoker classifications for each charge and then running our test statistic. We then take each test statistic and compare it to the value for the original data, using the values that are higher than the original to compute the p-value. After reproducing the original A/B test, we remove outliers from the data and then attempt to recreate the result to see if outliers affected the A/B test results. Finally, in our analysis, we will research the source of the data and evaluate it for statistical significance.

### Analysis

The test statistic in this analysis is the **absolute difference between the means of smoker and non-smoker charges**. That is, for two vectors of $n$ smoker charges $\mathbf{s}$ and $m$ non-smoker charges $\mathbf{\bar{s}}$, the test statistic is computed as

$$ \left | \frac{1}{n} \sum_{i=1}^n s_i - \frac{1}{m} \sum_{i=1}^m \bar{s}_i \right | $$

Then, to consider the dataset when outliers are not in the data, we arbitrarily consider outliers to be values that are more than 2.5 standard deviations ($\sigma$) away from the mean ($\mu$). That is, the set of outliers $O$ is

$$O = \{ c_i \mid c_i < \mu - 2.5 \sigma \} \cup \{ c_i \mid c_i > \mu + 2.5 \sigma \}$$

To examine the reliability of the data, we trace the data set back to its original source and evaluate its method of production and its original intended purpose. The data was retrieved from Kaggle for the original analysis, and because anyone can post a data set on Kaggle, it brings into question the validity of any conclusions drawn from the data.
